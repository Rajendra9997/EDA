# Handling_Outliers

```python
#segregating categorical and numerical variables
num_col=[fea for fea in data.columns if (data[fea].dtype !='O' and data[fea].dtype !=int)]
num_col
```

```python
X=data[num_col]
X['MonthlyIncome'].sum()
```

```python
# Taking Monthly Income as the variable
sns.boxplot(X['MonthlyIncome']) 
```

Method 1: Replacing Outliers in the dataset with threshold values or Winorization

```python
def find_boundaries(df,variable):
    IQR=df[variable].quantile(0.75)-df[variable].quantile(0.25)
    lower_boundary=df[variable].quantile(0.25)-(IQR*1.5)
    upper_boundary=df[variable].quantile(0.75)+(IQR*1.5)
    return upper_boundary,lower_boundary
```

```python
def replace_with_threshold(df,var):
     for variable in num_col:
        upper_boundary,lower_boundary=find_boundaries(df,variable)
        data.loc[data[variable]<lower_boundary,variable]=lower_boundary
        data.loc[data[variable]>upper_boundary,variable]=upper_boundary
```

```python
MonthlyIncome_upper_limit,MonthlyIncome_lower_limit=find_boundaries(X,'MonthlyIncome')
```

```python
MonthlyIncome_upper_limit,MonthlyIncome_lower_limit
```

```python
replace_with_threshold(X,'MonthlyIncome')
```

```python
X['MonthlyIncome'].sum()
```

Method 2: Trimming outliers from data set

```python
outliers_MonthlyIncome = np.where(X['MonthlyIncome'] >MonthlyIncome_upper_limit, True,
np.where(X['MonthlyIncome'] < MonthlyIncome_lower_limit,True, False))
```

```python
X_trimmed = X.loc[~outliers_MonthlyIncome] # deleting Outliers
X.shape, X_trimmed.shape
```

```python
sns.boxplot(y=X_trimmed['MonthlyIncome'])
```

```python
num_col=[fea for fea in data.columns if (data[fea].dtype !='O' and data[fea].dtype !=int)]
num_col
```

```python
X=data[num_col]
```

Method 1: Standardizing Features

Standardization is the process of centering the variable at zero and standizing the variance to 1.To
standardize the features we subtrsct the mean from each observation and then divide the result by
standard deviation:
z=(x-mean(x))/std(x)

```python
#Performing Mean Imputation
for var in num_col:
    value=X[var].mean()
    X[var]=X[var].fillna(value)
```

```python
# using sklearn
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled=scaler.transform(X)
X_scaled_final=pd.DataFrame(X_scaled)
X_scaled_final
```

Method 2: Performing mean normalization

In mean normaliztion ,we center the variable at zero and rescale the distribution to the value range.
This procedure involves subtracting from the mean from each observation and then dividing the
result by the difference between the minimum and maximum values.

x_scaled=(x-mean(x))/(max(x)-min(x))

The transformation results in a distribution centered around 0 ,with min and max values within
the range of -1 to 1

```python
from sklearn.preprocessing import StandardScaler,RobustScaler
scaler_mean=StandardScaler(with_mean=True,with_std=False)# No division by standard deviation
scaler_minmax=RobustScaler(with_centering=False,with_scaling=True,quantile_range=(0,100))
scaler_mean.fit(X)
scaler_minmax.fit(X)
X_scaled=scaler_minmax.transform(X)
X_scaled_final=pd.DataFrame(X_scaled)
X_scaled_final
```

Method 3: Scaling to the maximum and minimum values

Scaling to the minimum and minimum values squeezes the values of the variables between 0 and 1.
xscaled=(x-min(x))/(max(x)-min(x))

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X)
X_scaled=scaler.transform(X)
X_scaled_final=pd.DataFrame(X_scaled)
X_scaled_final
```

Method 4: Implementing Maximum absolute scaling

xscaled=x/max(x)

```python
from sklearn.preprocessing import MaxAbsScaler
scaler=MaxAbsScaler()
scaler.fit(X)
X_scaled=scaler.transform(X)
X_scaled_final=pd.DataFrame(X_scaled)
X_scaled_final
```

Method 5: Scaling with median and Quantiles

xscaled= (x-median(x))/(q3(x)-q1(x))

```python
scaler=RobustScaler()
scaler.fit(X)
X_scaled=scaler.transform(X)
X_scaled_final=pd.DataFrame(X_scaled)
X_scaled_final
```

Method 6: Scaling to vector unit length

xscaled =x/norm

where norm may be either Manhattan distance or Euclidean Distance

```python
from sklearn.preprocessing import Normalizer
scaler=Normalizer(norm='l2')# l2 for Euclidean Distance
scaler.fit(X)
X_scaled=scaler.transform(X)
X_scaled_final=pd.DataFrame(X_scaled)
X_scaled_final
```
